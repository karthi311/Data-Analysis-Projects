## Step 5 : Ask a Question about the data
    * Try to answer a question you have about the data using a plot or statistics

### 1.What are the locations with the fastest roller coasters(minimum of 10)?

```python
ax = df.query('Location != "Other"').groupby('Location')['Speed_mph'].agg(['mean','count']).query('count >= 10').sort_values('mean')['mean']\
.plot(kind='barh', figsize=(12,5), title = 'Average coast speed by Location')

ax.set_xlabel('Average Coaster Speed')
plt.show()
```

### 2.Which manufacturer has the most coasters?

```python
# Count coasters by manufacturer
manufacturer_counts = df['Manufacturer'].value_counts()

# Plot bar chart
plt.figure(figsize=(15, 15))
sns.barplot(x=manufacturer_counts.index, y=manufacturer_counts.values)
plt.title('Coaster Count by Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

### 3.What is the relationship between coaster height and speed?

```python
plt.scatter(df['Height_ft'], df['Speed_mph'])
plt.title('Coaster Height vs Speed')
plt.xlabel('Height (ft)')
plt.ylabel('Speed (mph)')
plt.show()
```

### 4.How has the number of inversions changed over time?

```python
# Group by year and count inversions
inversions_over_time = df.groupby('Year_Introduced')['Inversions'].mean().reset_index()

# Line plot
plt.plot(inversions_over_time['Year_Introduced'], inversions_over_time['Inversions'])
plt.title('Average Inversions Over Time')
plt.xlabel('Year Introduced')
plt.ylabel('Average Inversions')
plt.show()
```

### 5.What is the distribution of coaster types?

```python
# Count coaster types
type_counts = df['Type_Main'].value_counts()

# Plot pie chart
plt.figure(figsize=(8, 8))
sns.set_style('whitegrid')
plt.pie(type_counts.values, labels=type_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Coaster Types')
plt.show()
```
